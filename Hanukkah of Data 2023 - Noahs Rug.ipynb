```python
# We need to find the last name that matches the telephone number keypad.
# This is easier to do in Pandas / Python

# Create the telephone keypad
numbers_to_letters = {
    '2': ['a', 'b', 'c'],
    '3': ['d', 'e', 'f'],
    '4': ['g', 'h', 'i'],
    '5': ['j', 'k', 'l'],
    '6': ['m', 'n', 'o'],
    '7': ['p', 'q', 'r', 's'],
    '8': ['t', 'u', 'v'],
    '9': ['w', 'x', 'y', 'z']
}

letters_to_numbers = {letter: k for k, v in numbers_to_letters.items() for letter in v}

def letters_to_number(name):
    return "".join(letters_to_numbers[l] for l in name.lower())
    
def filter(x):
    last_name = x['name'].split()[-1]
    if (len(last_name) == 10 or len(last_name) == 7):
        area_code, a, b = x['phone'].split('-')
        num = letters_to_number(last_name)
        if (num == a + b or num == area_code + a + b):
            return True
    return False
    
customers[customers.apply(filter, axis=1)]
# print(letters_to_number('Hello'))
```

```python
# Let's find the cheap cousin

# Calculate the profit on every customer by summing the cost to Noah
# (unit_price - wholesale_cost) times the number of items purchased in that order (qty)

# We don't have to do a subquery because orders_items 

duckdb.sql("""
SELECT
name,
phone,
SUM((unit_price - wholesale_cost) * qty) as profit
FROM orders_items
JOIN products ON (orders_items.sku = products.sku)
JOIN orders ON (orders_items.orderid = orders.orderid)
JOIN customers ON (orders.customerid = customers.customerid)
GROUP BY name, phone
ORDER BY profit asc
LIMIT 1
""").df()
```

```python
# Find all of her orders for Noah's merch (SKU starts with COL)
duckdb.sql("""
SELECT
customers.*, orders.*, orders_items.qty, orders_items.unit_price, products.*
FROM orders
JOIN customers ON (orders.customerid = customers.customerid)
JOIN orders_items ON (orders.orderid = orders_items.orderid)
JOIN products ON (orders_items.sku = products.sku)
WHERE
    phone = '585-838-9161'
    AND products.sku LIKE 'COL%'
""").df()
```

```python
# Then find orders that happened around the same time for a similarly named item (i.e. without the color in parens)

# The honest-to-God is I just looked up each of the 4 orders above and found similar orders.
# But this is sql-golf, right? Where's the fun in one-by-one searches?

duckdb.sql("""
WITH all_orders AS (
    SELECT
        phone,
        ordered,
        products.desc AS named,
        UNNEST(regexp_extract(named, '([^\(]+) \((\w+)\)', ['bare_name', 'color']))
    FROM orders
    JOIN customers ON (orders.customerid = customers.customerid)
    JOIN orders_items ON (orders.orderid = orders_items.orderid)
    JOIN products ON (orders_items.sku = products.sku)
    WHERE products.sku LIKE 'COL%'
),

sherri_orders AS (
    SELECT *
    FROM all_orders
    WHERE phone = '585-838-9161'
)

SELECT
    all_orders.bare_name,
    all_orders.color AS other_color,
    sherri_orders.color AS sherri_color,
    sherri_orders.ordered AS sherri_ordered,
    all_orders.ordered AS other_ordered,
    all_orders.phone,
    date_diff('hour', sherri_ordered, other_ordered) AS time_diff
FROM sherri_orders
LEFT OUTER JOIN all_orders ON (all_orders.bare_name = sherri_orders.bare_name)
WHERE time_diff = 0
    AND other_color != sherri_color
""").df()
```

```python
# We are looking for a collector of Noahs items. We can build on our previous query to find
# out 1) how many items does Noah sell in total, minus colors, and 2) who has purchased the most
# of these items?

duckdb.sql("""
WITH all_orders AS (
    SELECT
        phone,
        ordered,
        products.desc AS named,
        UNNEST(regexp_extract(named, '([^\(]+) \((\w+)\)', ['bare_name', 'color']))
    FROM orders
    JOIN customers ON (orders.customerid = customers.customerid)
    JOIN orders_items ON (orders.orderid = orders_items.orderid)
    JOIN products ON (orders_items.sku = products.sku)
    WHERE products.sku LIKE 'COL%'
),


-- Find the total number of Noah's items 

SELECT COUNT (DISTINCT bare_name)
FROM all_orders
""").df()
```

```python
# With the knowledge that there are 8 items total, group all item totals by the number of items purchased.
# For sanity we can verify that there is only one person with all 8 items (ORDER BY item_count DESC)

duckdb.sql("""
WITH all_orders AS (
    SELECT
        phone,
        ordered,
        products.desc AS named,
        UNNEST(regexp_extract(named, '([^\(]+) \((\w+)\)', ['bare_name', 'color']))
    FROM orders
    JOIN customers ON (orders.customerid = customers.customerid)
    JOIN orders_items ON (orders.orderid = orders_items.orderid)
    JOIN products ON (orders_items.sku = products.sku)
    WHERE products.sku LIKE 'COL%'
)

SELECT *
FROM (
    SELECT phone, COUNT (DISTINCT bare_name) as item_count
    FROM all_orders
    GROUP BY phone
)
WHERE item_count = 8
-- ORDER BY item_count DESC
""").df()
```
