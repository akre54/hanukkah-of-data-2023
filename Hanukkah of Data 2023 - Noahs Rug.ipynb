```python
import duckdb

duckdb.query("CREATE TABLE customers AS SELECT * FROM 'noahs-customers.csv'")
customers = duckdb.sql("SELECT * FROM customers").df()
duckdb.query("CREATE TABLE orders AS SELECT * FROM 'noahs-orders.csv'")
orders = duckdb.sql("SELECT * FROM orders").df()
duckdb.query("CREATE TABLE orders_items AS SELECT * FROM 'noahs-orders_items.csv'")
orders_items = duckdb.sql("SELECT * FROM orders_items").df()
duckdb.query("CREATE TABLE products AS SELECT * FROM 'noahs-products.csv'")
products = duckdb.sql("SELECT * FROM products").df()

combined = duckdb.sql("""
SELECT orders.*, customers.*, orders_items.qty, orders_items.unit_price, products.*
FROM orders
JOIN customers ON (orders.customerid = customers.customerid)
JOIN orders_items ON (orders.orderid = orders_items.orderid)
JOIN products ON (orders_items.sku = products.sku)
""").df()
combined
```

# Day 1

```python
# We need to find the last name that matches the telephone number keypad.
# This is easier to do in Pandas / Python

# Create the telephone keypad
numbers_to_letters = {
    '2': ['a', 'b', 'c'],
    '3': ['d', 'e', 'f'],
    '4': ['g', 'h', 'i'],
    '5': ['j', 'k', 'l'],
    '6': ['m', 'n', 'o'],
    '7': ['p', 'q', 'r', 's'],
    '8': ['t', 'u', 'v'],
    '9': ['w', 'x', 'y', 'z']
}

letters_to_numbers = {letter: k for k, v in numbers_to_letters.items() for letter in v}

def letters_to_number(name):
    return "".join(letters_to_numbers[l] for l in name.lower())
    
def filter(x):
    last_name = x['name'].split()[-1]
    if (len(last_name) == 10 or len(last_name) == 7):
        area_code, a, b = x['phone'].split('-')
        num = letters_to_number(last_name)
        if (num == a + b or num == area_code + a + b):
            return True
    return False
    
customers[customers.apply(filter, axis=1)]
# print(letters_to_number('Hello'))
```

```python
# BKY appears to be bakery, DLI is deli, PET is pet, TOY is toys, HOM is home products.
# Can't tell what COL is but seems to be something about specially colored objects?
# We're looking for cleaning supplies, so lets look in Home with the rest of the conditions (J.P. and 2017) added

combined[
    (combined['ordered'].dt.to_period('Y') == '2017') &
    (combined['name'].str.match('^J\w+ P\w+')) &
    (combined['sku'].str.startswith('HOM'))
]
```

```python
# Or as a sql query:
duckdb.sql("""
SELECT customers.*
FROM orders
JOIN customers ON (orders.customerid = customers.customerid)
JOIN orders_items ON (orders.orderid = orders_items.orderid)
JOIN products ON (orders_items.sku = products.sku)
WHERE date_part('year', CAST(orders.ordered AS DATE)) = 2017
AND customers.name GLOB 'J* P*'
AND products.sku GLOB 'HOM*'
""")
```

```python
# This one can be done in pure sql, so lets give that a shot.
# First off, list out our conditions. We know that they person was born in
# the Year of the Rabbit and are a Cancer, so a quick Google brings up those dates.
# Finally, lets filter results to the same neighborhood as Joshua Peterson, Jamaica, NY

year_of_the_rabbit_years = '1927, 1939, 1951, 1963, 1975, 1987, 1999, 2011, 2023'
cancer_dates = 'June 21 - July 22'

query = """
SELECT *, CAST(customers.birthdate AS DATE) as birthdate
FROM customers
WHERE date_part('year', birthdate) IN ({0})
AND ((date_part('month', birthdate) = 6 AND date_part('day', birthdate) > 21)
OR (date_part('month', birthdate) = 7 AND date_part('day', birthdate) < 22))
AND citystatezip LIKE '%Jamaica%'
"""

duckdb.sql(query.format(year_of_the_rabbit_years)).df()
```

```python
# We know that the person ordered pastries (category BKY) before 5 AM, and multiple pastries (qty > 1)
# The text says it was weeks later, so we know it's sometime after 2017, but there aren't many good candiates.
# We can also use the birthdate and gender of the name because she was on Tinder to guess that she is
# close to Robert in age

duckdb.sql("""
SELECT
substring(products.sku, 0, 4) AS category,
extract('hour' from ordered) AS hour,
extract('year' from ordered) AS year,
customers.*, orders.*, orders_items.qty, orders_items.unit_price, products.*
FROM orders
JOIN customers ON (orders.customerid = customers.customerid)
JOIN orders_items ON (orders.orderid = orders_items.orderid)
JOIN products ON (orders_items.sku = products.sku)
WHERE category = 'BKY'
AND HOUR > 2
AND HOUR < 5
AND qty > 1
""").df()
```
